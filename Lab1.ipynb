#Введение

В данной лабораторной работе требуется реализовать исскуственные нейроны с различными функциями активации:


*   Тождественная функция
*   Единичная ступенька
* Логистическая (сигмоида, гладкая ступенька)
* Гиперболический тангенс (th)
* Арктангенс
* Softsign



#Реализация

Импортируем нужные библиотеки

```python
import numpy as np
```

Реализуем класс искусственного нейрона

```python
class Neuron:
    def __init__(self, weights, bias):
        self.weights = np.array(weights)
        self.bias = bias

    def activate(self, inputs, activation_type='step'):
        z = np.dot(self.weights, inputs) + self.bias

        if activation_type == 'identity':
            return z
        elif activation_type == 'step':
            return 1 if z >= 0 else 0
        elif activation_type == 'sigmoid':
            return 1 / (1 + np.exp(-z))
        elif activation_type == 'tanh':
            return np.tanh(z)
        elif activation_type == 'arctan':
            return np.arctan(z)
        elif activation_type == 'softsign':
            return z / (1 + abs(z))
        else:
            return z
```

Демонстрация работы нейрона с различными функциями активации

```python
weights = [0.6, -0.4]
bias = 0.3
inputs = [1, 0]

neuron = Neuron(weights, bias)

print(f"Веса: {weights}")
print(f"Входы: {inputs}")
print(f"Смещение: {bias}")

print(f"\nРезультаты:")
print(f"Тождественная: {neuron.activate(inputs, 'identity')}")
print(f"Ступенчатая:   {neuron.activate(inputs, 'step')}")
print(f"Сигмоида:      {neuron.activate(inputs, 'sigmoid')}")
print(f"Tanh:          {neuron.activate(inputs, 'tanh')}")
print(f"Арктангенс:    {neuron.activate(inputs, 'arctan')}")
print(f"Softsign:      {neuron.activate(inputs, 'softsign')}")
```
